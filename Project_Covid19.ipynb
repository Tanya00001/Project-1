#### Surely, at higher rate..

```python
# last day cases, in the whole world

Last = cov19[cov19["date"] == "2020-05-24"]
```

```python
Last
```

```python
# Sorting this dataset

Max = Last.sort_values(by = "total_cases", ascending = False)
Max
```

```python
# Top 5 countries with maximum cases

Top_5 = Max[1:6]
Top_5
```

```python
sns.barplot(x = "location", y = "total_cases", data = Top_5, hue = "location")
```

```python
Top = cov19[(cov19["location"] == "United States") | (cov19["location"] == "Brazil") | (cov19["location"] == "Russia") | (cov19["location"] == "Italy") | (cov19["location"] == "Germany")]
```

```python
sns.barplot(x = "location", y = "total_cases", data = Top, hue = "date")
sns.set(rc = {'figure.figsize': (10,10)})
```

#### _Here, we can understand the cases of Top 5 countries._

```python
# Understanding cases of India and China.

I_C = cov19[(cov19["location"] == "India") | (cov19["location"] == "China")]
```

```python
sns.barplot(x = "location", y = "total_cases", data = I_C, hue = "date")
sns.set(rc = {'figure.figsize': (20,20)})
```

#### If we will look at the BarPlot, in China, cases were increasing from the first day and cases were higher and higher because the Covid-19 first identified at Wuhan City, China, from there it began spreading all over the World. But, in India, cases were few in the beginning, and gradually started increasing.

```python
# For the sake of ease, dropping NaN values.

data1 = cov19.dropna()
data1
```

```python
# Some countries data are available, those having prevalence of diabetes.


P = sns.barplot(x = "location", y = "diabetes_prevalence", data = data1)
P.set_xticklabels(P.get_xticklabels(), rotation = 40);
```

#### _China's data is not available in this dataset, but China is the country with highest number of diabetics worldwide, with around 116 million people suffering from the disease._



### According to WHO(World Health Organization), diabetes is the leading cause of death in Mexico. We can observe this thing in the above BarPlot.

```python
# Countries having more people aged 70 older.

P = sns.barplot(x = "location", y = "aged_70_older", data = data1)
P.set_xticklabels(P.get_xticklabels(), rotation = 40);
```
